Общата вариация в данните се разлага на компоненти, дължащи се на първия фактор, втория фактор и случайна вариация (остатък). Отхвърлянето на нулевата хипотеза при избрано ниво на значимост $\alpha=0.05$ означава статистически значимо влияние на съответния фактор върху зависимата променлива. Анализът помага да се установи дали наблюдаваните разлики са систематични или случайни.


# Решаване на задачите със Python

## Задача 1.

```python
data = {'V': [35, 50, 65, 80, 95, 110],
        'W': [20, 30, 40, 50, 60, 70],
        'S': [16, 26, 41, 62, 88, 119]
        }
df = pd.DataFrame(data)

# Подготовка на матрицата
X = df[['V', 'W']]
X = sm.add_constant(X)
y = df['S']

model_full = sm.OLS(y, X).fit()

print(model_full.summary())

a, b = np.polyfit(df['V'], df['W'], 1)

print(f"--- МАТЕМАТИЧЕСКА ВРЪЗКА МЕЖДУ V И W ---")
print(f"Функцията е: W = {a:.4f} * V + ({b:.4f})")
print(f"Това съответства на теоретичната зависимост: W = (2/3)*V - 10/3")

plt.figure(figsize=(6, 4))
plt.scatter(df['V'], df['W'], color='blue', label='Данни (V, W)')
plt.plot(df['V'], a * df['V'] + b, color='red', linestyle='--', label=f'Линейна функция: W={a:.2f}V {b:.2f}')
plt.title("Визуално доказателство за линейна зависимост")
plt.xlabel("Скорост (V)")
plt.ylabel("Влажност (W)")
plt.legend()
plt.grid(True)
plt.show()
```

Стандартна формула:
$$ S = \beta_0 + \beta_1 \cdot V + \beta_2 \cdot W $$

Имаме зависимост, където:

$$ W = \frac{2}{3}V - \frac{10}{3} $$

От там следва, че:

$$ \hat{S} = -4.4778 - 5.9368 \cdot V + 10.9681 \cdot \left(\frac{2}{3}V - \frac{10}{3}\right) $$
$$ \hat{S} = -4.4778 - 5.9368 \cdot V + 7.3121 \cdot V - 36.5603 $$
$$ \hat{S} = -41.0381 + 1.3753 \cdot V $$

Което значи че влажността не добавя нова информация, затова моделът може да не я използва

```python
data = {
    'V': [35, 50, 65, 80, 95, 110],
    'W': [20, 30, 40, 50, 60, 70],
    'S': [16, 26, 41, 62, 88, 119]
}
df = pd.DataFrame(data)

X = df[['V', 'W']]
X = sm.add_constant(X)
y = df['S']

# Проверка за мултиколинеарност (VIF)
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

# Избор на модел
if vif_data[vif_data["feature"] == "W"]["VIF"].values[0] > 10:
    print("Открита е зависимост (мултиколинеарност). Използваме само скорост (V).")
    X_final = X[['const', 'V']]
else:
    print("Данните са независими. Използваме пълния модел с V и W.")
    X_final = X

model = sm.OLS(y, X_final).fit()
print(model.summary())

prediction = model.predict([1, 70])
print(f"\nПрогноза за 70 км/ч: {prediction[0]:.3f} метра")
```

### Разсъждение и обобщение:
Поради перфектната линейна зависимост между скоростта $(V)$ и влажността $(W)$, данните проявяват мултиколинеарност, което прави едновременното им включване математически нестабилно. Поради това е приложен еднофакторен модел:
$ \hat{S}=-41.04+1.38\cdot V $
Изводи:
- Адекватност: Моделът е статистически значим $(p < 0.001)$ и обяснява $96.6%$ от вариацията на спирачния път $(R^2=0.966)$
- Интерпретация: С всяко увеличение на скоростта с 1 км/ч, спирачният път се увеличава средно с 1.38 м.
- Прогноза: При 70 км/ч очакваният спирачен път е 55.23 м.


## Задача 2.

```python
data = {
    'Вид': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Склад': ['I', 'II', 'III', 'IV', 'I', 'II', 'III', 'IV', 'I', 'II', 'III', 'IV'],
    'Годност': [33.1, 39.0, 33.8, 33.0, 27.2, 36.9, 29.7, 28.5, 27.4, 29.2, 26.7, 30.9]
}

df = pd.DataFrame(data)
print(df)

# Изпълнение на ANOVA
model = ols('Годност ~ C(Вид) + C(Склад)', data=df).fit() # Важно: добавихме C(), за да сме сигурни, че ги третира като категории
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nANOVA таблица:")
print(anova_table)
```

### Разсъждение и обобщение:
Извършен е двуфакторен ANOVA анализ без взаимодействие за определяне влиянието на „Вид консерва“ и „Тип склад“ върху годността на продуктите.

Изводи:
- Вид консерва: Отхвърляме нулевата хипотеза $(p = 0.0237 < 0.05)$, което доказва, че видът на консервата има статистически значим ефект върху срока на годност.
- Тип склад: Нулевата хипотеза не се отхвърля $(p = 0.0791 > 0.05)$, което означава, че типът склад не оказва съществено влияние върху годността при избраното ниво на значимост.
Обобщено, видът на консервата е основният детерминиращ фактор за нейната годност в настоящото изследване.
